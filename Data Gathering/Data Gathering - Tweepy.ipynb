```python
#imports for tweepy and pandas for showing the data

import tweepy
import pandas as pd
import csv
```

```python
#getting authentication to use the twitter api

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, wait_on_rate_limit=True)
```

```python
#declare data containers and header names

header_names = ['Data of Tweet', 'Tweet Content']
tweet_list = []
tweets_all = "C:/Users/Kirby Wenceslao/Desktop/Thesis/Data Gathering/tweets_dataset_ALL.csv"
```

```python
#Gathering Tweets by specified tweets

#1. October 08, 2022 Last day of filing of COC

label = "#label of dev account"

#df
tweets_1 = []

q = '(leni robredo OR bongbong marcos OR isko moreno domagoso OR manny pacman pacquiao OR ping lacson OR ernie abella OR leody de guzman OR norberto gonzales OR jose montemayor jr OR faisal mangondato) -is:retweet'
q2 = '(leni robredo OR bongbong marcos OR isko moreno domagoso OR manny pacman pacquiao OR ping lacson OR leody de guzman)'

start_time= '202110080000'
end_time= '202110090000'

tweets = tweepy.Cursor(api.search_full_archive, 
                       label = label, 
                       query = q, 
                       fromDate = start_time, 
                       toDate = end_time,
                       maxResults = 500).items(2000)

#add gathered tweets on master list of data
with open(tweets_all, mode='a', encoding='utf-8', newline='') as f:
    for tweet in tweets:
        writer = csv.writer(f)
        if hasattr(tweet, 'extended_tweet'):
            writer.writerow([tweet.created_at, tweet.extended_tweet['full_text']])
            tweet_list.append([tweet.created_at, tweet.extended_tweet['full_text']])
            tweets_1.append([tweet.created_at, tweet.extended_tweet['full_text']]) 
        else:
            writer.writerow([tweet.created_at, tweet.text])
            tweet_list.append([tweet.created_at, tweet.text])
            tweets_1.append([tweet.created_at, tweet.text]) 
        
```

```python
#display current data, and create csv for specific date

df = pd.DataFrame(tweets_1, columns = header_names)
df.to_csv('tweets_dataset_1.csv', index = False)
df
```
